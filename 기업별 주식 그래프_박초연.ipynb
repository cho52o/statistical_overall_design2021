#### 날짜 추출

```python
# 날짜로 추출
# 20200323 20210107"
# 20200323 20210107

# 이전 일주일

data1 = data[data["일자"]>=20210101]
data2 = data1[data1["일자"]<=20210107]
```

#### NALL 값 확인

```python
data2["키워드"].isnull().sum()
```

```python
data2.head(2)
```

#### '키워드'열의 단어 추출

```python
df = pd.DataFrame()
for i in data2.index:
  df = df.append(pd.DataFrame(data2['키워드'][i].split(","), columns=["키워드"]), ignore_index = True)
df.count()
```

```python
df.head(2)
```

```python
df.tail(2)
```

#### 단어 빈도수 계산

```python
df_bin = df.groupby(['키워드']).size().reset_index()
df_bin.rename(columns = {0: '빈도'}, inplace = True)
```

```python
df_bin.head()
```

```python
df_bin.빈도.sort_values().tail(20)
```

#### 빈도수 내림차순 정렬

```python
df_bin_sort = df_bin.sort_values(by=['빈도'], axis=0, ascending=False)
```
